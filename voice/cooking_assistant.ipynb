Define *who* your assistant is. For a cooking assistant, context is key.

### 3.3 SSML & Pronunciation (The Secret Sauce)
Cooking terms can be tricky. We use **SSML (Speech Synthesis Markup Language)** to control pronunciation and pacing.
When using Twilio's Text-to-Speech options (like Amazon Polly or Google), SSML is a powerful tool.

More details here: https://www.twilio.com/en-us/changelog/conversationrelay-now-supports-ssml-tags-to-fine-tune-speech

````python
# Example of SSML to correct pronunciation of 'Sauté' and add pauses
ssml_text = """
<speak>
  Okay, first, let's <say-as interpret-as="characters">sauté</say-as> the onions. 
  <break time="0.5s"/> 
  Wait until they are golden brown.
</speak>
"""

print("Prepared SSML for synthesis.")
````
