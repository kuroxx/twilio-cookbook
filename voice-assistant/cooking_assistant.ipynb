## 1. Mise en place (Setup)

First, let's gather our ingredients. You'll need your Twilio Account SID and Auth Token.

```python
# Install necessary ingredients
%pip install twilio python-dotenv
```

```python
import os
from dotenv import load_dotenv
from twilio.rest import Client

# Load environment variables from a .env file
load_dotenv()

# Twilio Credentials
account_sid = os.environ.get('TWILIO_ACCOUNT_SID')
auth_token = os.environ.get('TWILIO_AUTH_TOKEN')

# Initialize the Twilio Client
client = Client(account_sid, auth_token)

print("✅ Ingredients loaded!")
```

## 3. Main Course: Voice Configuration 🗣️

The soul of our assistant is its voice. We want something warm, encouraging, and clear.

### A. Prompt Engineering (The Persona)
Define *who* your assistant is. For a cooking assistant, context is key.

```python
system_prompt = """
You are 'Chef Lumière', a warm, encouraging, and patient culinary assistant. 
Your goal is to help home cooks navigate recipes with ease. 
Keep instructions simple, concise, and safe.
If a user seems stressed, offer a calming reassured remark.
"""
print(f"Defined Persona:\n{system_prompt}")
```

### B. SSML & Pronunciation (The Secret Sauce)
Cooking terms can be tricky. We use **SSML (Speech Synthesis Markup Language)** to control pronunciation and pacing.
When using Twilio's Text-to-Speech options (like Amazon Polly or Google), SSML is a powerful tool.

```python
# Example of SSML to correct pronunciation of 'Sauté' and add pauses
ssml_text = """
<speak>
  Okay, first, let's <say-as interpret-as="characters">sauté</say-as> the onions. 
  <break time="0.5s"/> 
  Wait until they are golden brown.
</speak>
"""

print("Prepared SSML for synthesis.")
```

## 4. Side Dish: Webhook & ConversationRelay 📡

Now we need to connect our phone number to our logic. We use Twilio's **ConversationRelay** (or Media Streams) to handle the real-time audio flow.

1.  **Expose your local server**: Use `ngrok` to create a public URL (e.g., `https://my-tunnel.ngrok.io`).
2.  **Configure the properties**: Point the voice webhook to your application.

```python
# This is a representation of the TwiML we will serve to Twilio
webhook_url = "wss://my-tunnel.ngrok.io/media-stream"

twiml_instruction = f"""
<Response>
    <Connect>
        <Stream url="{webhook_url}" />
    </Connect>
</Response>
"""

print("TwiML Configuration Ready:")
print(twiml_instruction)
```
